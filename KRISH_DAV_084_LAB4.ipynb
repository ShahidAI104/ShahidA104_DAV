```python
import pandas as pd
import numpy as np


csv_data = """COUNTRY,YEAR,POPULATION,GDP,LIFE
USA,2018,327000000,20513000,78.7
USA,2019,328200000,21433000,78.8
USA,2020,331000000,20936000,78.9
GERMANY,2018,83000000,3996000,81.0
GERMANY,2019,83100000,3861000,81.1
GERMANY,2020,83200000,3806000,81.2
FRANCE,2018,67000000,2778000,82.5
FRANCE,2019,67200000,2715000,82.6
FRANCE,2020,67400000,2603000,82.7
"""

with open('DAV DATA.csv', 'w') as file:
    file.write(csv_data)

df = pd.read_csv('DAV DATA.csv')


print("Original DataFrame:")
print(df)


print("\nMean Population:", df['POPULATION'].mean())
print("Sum of GDP:", df['GDP'].sum())
print("Max Life Expectancy:", df['LIFE'].max())
print("Min GDP:", df['GDP'].min())
print("Mean GDP (using numpy):", np.mean(df['GDP']))
print("Standard Deviation of Population:", np.std(df['POPULATION']))
print("Sum of Population:", np.sum(df['POPULATION']))


df['GDP'] = df['GDP'] * 1.10
print("\nDataFrame after increasing GDP by 10%:")
print(df)


df_multi = df.set_index(['COUNTRY', 'YEAR'])
print("\nMultiIndex DataFrame:")
print(df_multi)


df_swapped = df_multi.swaplevel('COUNTRY', 'YEAR').sort_index()
print("\nSwapped index levels (YEAR, COUNTRY) sorted:")
print(df_swapped)

df_unstacked = df_multi.unstack(level='YEAR')
print("\nUnstacked DataFrame (YEAR as columns):")
print(df_unstacked)


print("\nPopulation of USA over the years:")
print(df_multi.loc['USA', 'POPULATION'])

print("\nPopulation of GERMANY over the years:")
print(df_multi.loc['GERMANY', 'POPULATION'])
```
