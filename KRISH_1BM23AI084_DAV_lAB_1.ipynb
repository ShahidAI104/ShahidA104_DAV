```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

df = pd.read_csv('Employee_data.csv')

display("Missing Values:\n", df.isnull())
display("\nNot Missing Values:\n", df.notnull())
```

```python
occupation_mode = df['Occupation'].mode()[0]
df['Occupation'] = df['Occupation'].fillna(occupation_mode)

satisfaction_mean = df['Satisfaction_Level'].mean()
df['Satisfaction_Level'] = df['Satisfaction_Level'].fillna(satisfaction_mean)

def satisfaction_to_binary(x):
    return 'High' if x > 0.7 else 'Low'

df['Satisfaction_Binary'] = df['Satisfaction_Level'].apply(satisfaction_to_binary)

purchase_map = {'High': 2, 'Medium': 1, 'Low': 0}
df['Purchase_History_Num'] = df['Purchase_History'].map(purchase_map)
```

```python
df['Income_Zscore'] = zscore(df['Income'].dropna())
outliers_zscore = df[(df['Income_Zscore'] > 3) | (df['Income_Zscore'] < -3)]

Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = df[(df['Income'] < (Q1 - 1.5 * IQR)) | (df['Income'] > (Q3 + 1.5 * IQR))]

display("Outliers using Z-score method:\n", outliers_zscore[['ID', 'Income', 'Income_Zscore']])
display("\nOutliers using IQR method:\n", outliers_iqr[['ID', 'Income']])

print(f"Missing Years_Employed count: {df['Years_Employed'].isnull().sum()}")

median_years = df['Years_Employed'].median()
df['Years_Employed_Filled'] = df['Years_Employed'].fillna(median_years)

display(df.head())
```
